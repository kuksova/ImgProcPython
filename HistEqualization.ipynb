```python
from skimage.io import imread, imsave, imshow
%matplotlib inline
from skimage import img_as_float, img_as_ubyte
import numpy as np
img = imread('landscape.png')
pix = img.shape[0]*img.shape[1]

def cdf(x, values):
    count = 0
    m = values[0]
    for i in range(x):
        count += values[i]
        if values[i]<m:
            m = values[i] 
    return (count, m)

values, bin_edges = np.histogram(img, bins=range(257))
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        a = img[i, j]
        count, m = cdf(a, values)
        img[i,j] = round(((count - m)/pix-1)*255)

imshow(img)
```
