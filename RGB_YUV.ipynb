```python
from skimage.io import imread, imsave, imshow
from skimage import img_as_float, img_as_ubyte
%matplotlib inline
import numpy as np
img = imread('tiger-color.png')
img_f = img_as_float(img)
Y =  0.2126*img_f[:, :, 0]+0.7152*img_f[:, :, 1]+0.0722*img_f[:, :, 2]
U = (-1)*0.0999*img_f[:, :, 0] - 0.3360*img_f[:, :, 1]+0.4360*img_f[:, :, 2]
V = 0.6150*img_f[:, :, 0] - 0.5586*img_f[:, :, 1]-0.0563*img_f[:, :, 2]

#from numpy import sort
#y_mod = sort(y, axis=None)
#k = round(y_mod.size*0.05)
#x_min, x_max = y_mod[k], y_mod[-k-1]

#Y_sorted = np.sort(Y, axis=None)
#k = int(np.round(Y_sorted.size * 0.05))
#xmin = Y_sorted[k]
#xmax = Y_sorted[-k-1]
k = int(np.round(Y.size * 0.05))
img_sorted = np.sort(Y.ravel())
xmin = img_sorted[k]
xmax = img_sorted[-k]
print(k)
print(img_sorted )
print(xmin, xmax)
Y_out = ((Y - xmin) * (1.0 / (xmax - xmin)))
Yclip = np.clip(Y_out,0.0,1.0)
R = Yclip+1.2803*V
R = np.clip(R,0.0,1.0)
G = Yclip-0.2148*U-0.3805*V
G = np.clip(G,0.0,1.0)
B = Yclip+2.1279*U
B = np.clip(B,0.0,1.0)
img_out = np.dstack((R,G,B))
imshow(img_as_ubyte(img_out))

```
